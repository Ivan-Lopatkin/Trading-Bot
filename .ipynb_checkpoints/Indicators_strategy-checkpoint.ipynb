```python
import os
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt

from alpha_vantage.timeseries import TimeSeries

import mplcyberpunk
plt.style.use('cyberpunk')
```

```python
def preprocessing(file_path):
    cols = ['DATE', 'TIME', 'OPEN', 'HIGH', 'LOW', 'CLOSE', 'VOL']
    data = pd.read_csv('data/Beijing airport_.csv', delimiter=';', header=None, names=cols)
    
    data.drop(['TIME'], axis=1, inplace=True)
    return data
```

```python
files = os.listdir(path='./data/')
```

```python
files
```

```python
data = preprocessing(files[0])
```

```python
data.tail()
```

```python
data.columns = ['OPEN', 'HIGH', 'LOW', 'CLOSE', 'VOL']

# 1. MACD
data['EMA12'] = data['CLOSE'].ewm(span=12, adjust=False).mean()
data['EMA26'] = data['CLOSE'].ewm(span=26, adjust=False).mean()
data['MACD'] = data['EMA12'] - data['EMA26']
data['Signal_Line'] = data['MACD'].ewm(span=9, adjust=False).mean()

# 2. RSI
window_length = 14
delta = data['CLOSE'].diff()
gain = (delta.where(delta > 0, 0)).rolling(window=window_length).mean()
loss = (-delta.where(delta < 0, 0)).rolling(window=window_length).mean()
rs = gain / loss
data['RSI'] = 100 - (100 / (1 + rs))

# 3. Полосы Боллинджера
data['Bollinger_Middle'] = data['CLOSE'].rolling(window=20).mean()
data['Bollinger_Upper'] = data['Bollinger_Middle'] + 2 * data['CLOSE'].rolling(window=20).std()
data['Bollinger_Lower'] = data['Bollinger_Middle'] - 2 * data['CLOSE'].rolling(window=20).std()
```

```python
# Определим количество дней 
test_days = data.shape[0]
window = 0  # Окно времени в днях для проверки сигналов RSI и Боллинджера
test_data = data[-test_days:].copy()

# Инициализация параметров портфеля
initial_balance = 100000  # начальный баланс в рублях
balance = initial_balance
shares = 0
portfolio_value = []
pending_buy = False
pending_sell = False

# Стратегия торговли на основе сигналов RSI, Боллинджера и подтверждения MACD
for i in range(window, len(test_data)):
    # Проверка условий для покупки (RSI и Боллинджер) в пределах окна времени
    rsi_buy_signal = (test_data['RSI'][i-window:i+1] <= 30).any()
    bollinger_buy_signal = (test_data['CLOSE'][i-window:i+1] <= test_data['Bollinger_Lower'][i-window:i+1]).any()

    # Проверка условий для продажи (RSI и Боллинджер) в пределах окна времени
    rsi_sell_signal = (test_data['RSI'][i-window:i+1] >= 70).any()
    bollinger_sell_signal = (test_data['CLOSE'][i-window:i+1] >= test_data['Bollinger_Upper'][i-window:i+1]).any()

    # Активируем отложенный ордер на покупку при выполнении условий RSI и Боллинджера
    if rsi_buy_signal and bollinger_buy_signal:
        pending_buy = True

    # Активируем отложенный ордер на продажу при выполнении условий RSI и Боллинджера
    if rsi_sell_signal and bollinger_sell_signal:
        pending_sell = True

    # Подтверждающий сигнал MACD для покупки
    if pending_buy and test_data['MACD'].iloc[i] > test_data['Signal_Line'].iloc[i] and test_data['MACD'].iloc[i-1] <= test_data['Signal_Line'].iloc[i-1]:
        shares_to_buy = balance // test_data['CLOSE'].iloc[i]
        balance -= shares_to_buy * test_data['CLOSE'].iloc[i]
        shares += shares_to_buy
        pending_buy = False  # Сбрасываем флаг покупки

    # Подтверждающий сигнал MACD для продажи
    if pending_sell and test_data['MACD'].iloc[i] < test_data['Signal_Line'].iloc[i] and test_data['MACD'].iloc[i-1] >= test_data['Signal_Line'].iloc[i-1]:
        balance += shares * test_data['CLOSE'].iloc[i]
        shares = 0
        pending_sell = False  # Сбрасываем флаг продажи

    # Обновляем значение портфеля
    portfolio_value.append(balance + shares * test_data['CLOSE'].iloc[i])
    
#     # Построим графики
#     plt.figure(figsize=(14, 4))
#     plt.plot(np.arange(0, len(test_data))[:i + 1], test_data['CLOSE'][:i + 1].values)
#     plt.title('Stock price')
#     plt.show()

#     # График стоимости портфеля
#     plt.figure(figsize=(14, 4))
#     plt.plot(np.arange(0, len(test_data))[:i + 1], portfolio_value, label='Portfolio Value', color='green')
#     plt.title('Portfolio Value Over Last 3 Months')
#     plt.xlabel('date')
#     plt.ylabel('Portfolio Value (RUB)')
#     plt.legend()
#     plt.show()

#     # График MACD
#     plt.figure(figsize=(14, 4))
#     plt.plot(np.arange(0, len(test_data))[:i + 1], test_data['MACD'].values[:i + 1], label='MACD', color='blue')
#     plt.plot(np.arange(0, len(test_data))[:i + 1], test_data['Signal_Line'].values[:i + 1], label='Signal Line', color='red')
#     plt.title('MACD Indicator')
#     plt.xlabel('date')
#     plt.ylabel('MACD Value')
#     plt.legend()
#     plt.show()

#     # График RSI
#     plt.figure(figsize=(14, 4))
#     plt.plot(np.arange(0, len(test_data))[:i + 1], test_data['RSI'].values[:i + 1], label='RSI', color='purple')
#     plt.axhline(70, color='red', linestyle='--', label='Overbought (70)')
#     plt.axhline(30, color='green', linestyle='--', label='Oversold (30)')
#     plt.title('RSI')
#     plt.xlabel('date')
#     plt.ylabel('RSI Value')
#     plt.legend()
#     plt.show()

#     # График полос Боллинджера
#     plt.figure(figsize=(14, 4))
#     plt.plot(np.arange(0, len(test_data))[:i + 1], test_data['CLOSE'].values[:i + 1], label='Close Price', color='black')
#     plt.plot(np.arange(0, len(test_data))[:i + 1], test_data['Bollinger_Upper'].values[:i + 1], label='Upper Bollinger Band', color='orange')
#     plt.plot(np.arange(0, len(test_data))[:i + 1], test_data['Bollinger_Lower'].values[:i + 1], label='Lower Bollinger Band', color='orange')
#     plt.fill_between(np.arange(0, len(test_data))[:i + 1], test_data['Bollinger_Lower'].values[:i + 1], test_data['Bollinger_Upper'].values[:i + 1], color='orange', alpha=0.1)
#     plt.title('Bollinger Bands')
#     plt.xlabel('date')
#     plt.ylabel('Price')
#     plt.legend()
#     plt.show()
    
# Добавим данные о стоимости портфеля в DataFrame
test_data = test_data.iloc[window:]
test_data['Portfolio_Value'] = portfolio_value
```

```python
# Построим графики
plt.figure(figsize=(14, 4))
plt.plot(np.arange(0, len(test_data)), test_data['CLOSE'].values)
plt.title('Stock price')
plt.show()

# График стоимости портфеля
plt.figure(figsize=(14, 4))
plt.plot(np.arange(0, len(test_data)), test_data['Portfolio_Value'].values, label='Portfolio Value', color='green')
plt.title('Portfolio Value Over Last 3 Months')
plt.xlabel('date')
plt.ylabel('Portfolio Value (RUB)')
plt.legend()
plt.show()

# График MACD
plt.figure(figsize=(14, 4))
plt.plot(np.arange(0, len(test_data)), test_data['MACD'].values, label='MACD', color='blue')
plt.plot(np.arange(0, len(test_data)), test_data['Signal_Line'].values, label='Signal Line', color='red')
plt.title('MACD Indicator')
plt.xlabel('date')
plt.ylabel('MACD Value')
plt.legend()
plt.show()

# График RSI
plt.figure(figsize=(14, 4))
plt.plot(np.arange(0, len(test_data)), test_data['RSI'].values, label='RSI', color='purple')
plt.axhline(70, color='red', linestyle='--', label='Overbought (70)')
plt.axhline(30, color='green', linestyle='--', label='Oversold (30)')
plt.title('RSI')
plt.xlabel('date')
plt.ylabel('RSI Value')
plt.legend()
plt.show()

# График полос Боллинджера
plt.figure(figsize=(14, 4))
plt.plot(np.arange(0, len(test_data)), test_data['CLOSE'].values, label='Close Price', color='black')
plt.plot(np.arange(0, len(test_data)), test_data['Bollinger_Upper'].values, label='Upper Bollinger Band', color='orange')
plt.plot(np.arange(0, len(test_data)), test_data['Bollinger_Lower'].values, label='Lower Bollinger Band', color='orange')
plt.fill_between(np.arange(0, len(test_data)), test_data['Bollinger_Lower'], test_data['Bollinger_Upper'], color='orange', alpha=0.1)
plt.title('Bollinger Bands')
plt.xlabel('date')
plt.ylabel('Price')
plt.legend()
plt.show()
```
